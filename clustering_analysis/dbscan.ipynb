```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=150,
                  n_features=2,
                  centers=3,
                  cluster_std=.5,
                  shuffle=True,
                  random_state=0)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],
            X[:,1],
            c='white',
            marker='o',
            edgecolor='black',
            s=50)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from matplotlib import cm
import numpy as np



db = DBSCAN(eps=0.6, min_samples=10, metric='euclidean')
y_db = db.fit_predict(X)

n_clusters = len(np.unique(y_db))

cmap = cm.get_cmap('jet')
colors = [cmap(j/n_clusters) for j in range(n_clusters)]

for i_cluster in range(len(np.unique(y_db))):
    plt.scatter(X[y_db == i_cluster,0],
                X[y_db == i_cluster,1],
                color=colors[i_cluster],
                marker='o',
                edgecolor='black',
                s=50)
    
# noise
i_cluster = -1
plt.scatter(X[y_db == i_cluster,0],
            X[y_db == i_cluster,1],
            color=colors[i_cluster],
            marker='x',
            edgecolor='black',
            s=50)
plt.grid()
plt.tight_layout()
plt.show()
```
